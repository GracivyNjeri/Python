3.*Data Cleaning*
Removed missing values and duplicate rows from the dataset.

```python
#check for missing values
print(online_retail.isnull().sum())
```

```python
#remove rows with missing values
online_retail= online_retail.dropna() 
```

```python
#check for duplicated rows
print('Duplicte Row :', online_retail.duplicated().sum())
```

```python
#remove duplicate rows
online_retail = online_retail.drop_duplicates()
online_retail
```

4.*EDA Explotary Data Analysis*

```python
# Visualize Sales Over Time
online_retail['InvoiceDate'] =pd.to_datetime(online_retail['InvoiceDate'])
online_retail['InvoiceMonth'] = online_retail['InvoiceDate'].dt.to_period('M')

Monthly_sales = online_retail.groupby('InvoiceMonth')['Quantity'].sum()
plt.figure(figsize = (10,6))
Monthly_sales.plot (kind= 'bar', color = 'skyblue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Quantity Solid')
plt.show()
```

*Finding the Most Selling Product*


```python
#group by product and sum the quantity sold
 
product_sales = online_retail.groupby('Description')['Quantity'].sum().sort_values(ascending=False)

#Find the most selling prodduct
most_selling_product = product_sales.idxmax()
total_quantity_sold= product_sales.max()

#visualize top N most Selling product

top_n= 10
top_product = product_sales.head(top_n)

#Plot the top N most selling products
plt.figure(figsize=(12,6))
top_product.plot(kind='bar', color = 'teal')
plt.title(f'Top {top_n} Most selling Products')
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.xtick(rotation = 45 , ha = 'right')
plt.show()

print(f"The most selling product is '{most_selling_product}' with a total quantity sold of {total_quantity_sold}.")
```

*Online purchases by country*

```python

online_retail['InvoiceDate']= pd.to_datetime(online_retail['InvoiceDate'])
online_retail['InvoiceMonth']= online_retail['InvoiceDate'].dt.to_period('M')

#investigate the number of invoices for each country
country_purchases =online_retail.groupby('Country')['Invoice'].nunique().sort_values(ascending=False)

top_n_countries = 10


#Plot the top N countries with the Most online Purchases
plt.figure(figsize=(12,6))
country_purchases.head(top_n_countries).plot(kind= 'bar', color ='skyblue')
plt.title(f'Top {top_n_countries} countries with the Most online Purchases')
plt.xlabel('Country')
plt.ylabel('NO. of Invoices')


#highlight most sales and and least purchases
most_sales_country = 'united Kingdom'
least_sales_country ='Switzerland'

plt.axvline(x=most_sales_country, color='green',linestyle= '--', label = f'Most Sales:{most_sales_country}')
plt.axvline(x=least_sales_country, color='red',linestyle= '--', label = f'least Sales:{least_sales_country}')
plt.legend()
plt.show()





```
